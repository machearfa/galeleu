```python
import openai
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv
import base64 #transforma a imagem em numeros
```

```python
_ = load_dotenv(find_dotenv()) #procuramos a api no .env e carregamos no nosso ambiente
client = OpenAI()
```

```python
#criando a função de codificar a imagem em numeros
def encode_image(imagem):
    with open (imagem, "rb") as img:    #rb = read bit
        return base64.b64encode(img.read()).decode('utf-8') 
```

```python
caminho = "mensagem_do_tio.jpeg"
base_64_img = encode_image(caminho)
```

```python
texto = input("escreva a sua pergunta sobre a imagem: ")
```

```python
#configurando a resposta do meu modelo de ia, utilizando o metodo create
#definindo os argumentos model e messages (formato de dicionario,sendo o conteudo em lista de dict)
resposta = client.chat.completions.create(
    model='gpt-5.2',
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": texto},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{base_64_img}"
                    }
                }
            ]
        }
    ],
)          
```

```python
print(resposta)
```
