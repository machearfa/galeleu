```python
import openai
from dotenv import load_dotenv, find_dotenv

_ = load_dotenv(find_dotenv())
```

```python
#estamos instanciando a classe client no objeto client
client = openai.Client()
```

```python
arquivo = 'vitor_audio.mp3'
texto = 'Vinicius não quer fazer galileu'
```

```python
resposta = client.audio.speech.create(
    model='tts-1',
    voice='onyx',
    input=texto
)
```

```python
resposta.write_to_file(arquivo)
```
